```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample monthly sales data (units)
data = pd.DataFrame({
    'Category': ['Electronics', 'Clothing', 'Home Appliances', 'Groceries'],
    'Units_Sold': [350, 500, 200, 700]  # Replace with actual sales numbers
})

# Create a bar chart
plt.figure(figsize=(8,5))
bars = plt.bar(data['Category'], data['Units_Sold'], color=['blue', 'green', 'orange', 'purple'])

# Add titles and axis labels
plt.title("Monthly Units Sold by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Units Sold")
plt.xticks(rotation=15)

# Add data labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, yval, ha='center', va='bottom')

# Show the plot
plt.show()
```

Groceries – 700 units sold.
