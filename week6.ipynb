```python
import matplotlib.pyplot as plt
import numpy as np
weights = np.random.normal(loc=75, scale=15, size=100)  # mean=75kg, std dev=15kg
plt.hist(weights, bins=[40, 60, 80, 100, 120], color='skyblue', edgecolor='black')
plt.title('Weight Distribution of Gym Members')
plt.xlabel('Weight (kg)')
plt.ylabel('Number of Members')
plt.xticks([50, 70, 90, 110], ['Underweight', 'Normal', 'Overweight', 'Obese'])
plt.show()
```
