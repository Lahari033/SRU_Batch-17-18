```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Simple dataset
data = {
    'Courier': ['BlueDart', 'BlueDart', 'BlueDart',
                'Delhivery', 'Delhivery', 'Delhivery',
                'DTDC', 'DTDC', 'DTDC'],
    'Delivery_Time': [2, 3, 2, 4, 5, 3, 3, 2, 4]
}

df = pd.DataFrame(data)

# Create boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x='Courier', y='Delivery_Time', data=df)

# Add title and labels
plt.title('Comparison of Delivery Times by Courier')
plt.xlabel('Courier Service')
plt.ylabel('Delivery Time (Days)')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Courier': ['BlueDart', 'BlueDart', 'BlueDart',
                'Delhivery', 'Delhivery', 'Delhivery',
                'DTDC', 'DTDC', 'DTDC'],
    'Delivery_Time': [2, 3, 2, 4, 5, 3, 3, 2, 4],
    'Region': ['North', 'South', 'East',
               'North', 'South', 'East',
               'North', 'South', 'East']
}
df = pd.DataFrame(data)
plt.figure(figsize=(10,5))
sns.boxplot(x='Courier', y='Delivery_Time', hue='Region', data=df, palette='deep')
plt.title('Comparison of Delivery Times by Courier and Region')
plt.xlabel('Courier ')
plt.ylabel('Delivery Time ')
plt.legend(title='Region')
plt.show()
```
