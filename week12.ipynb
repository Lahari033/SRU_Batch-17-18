```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data: Average monthly temperature and rainfall

data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Temperature': [5, 7, 12, 18, 24, 28, 30, 29, 25, 18, 10, 6],
    'Rainfall': [80, 60, 70, 55, 45, 40, 35, 40, 50, 65, 75, 85]
})

# Melt data for easy plotting with seaborn
data_melted = data.melt(id_vars='Month', value_vars=['Temperature', 'Rainfall'],
                        var_name='Variable', value_name='Value')

# Plotting using Seaborn lineplot

sns.set_style("ticks")  # Aesthetic style
plt.figure(figsize=(10, 6))

sns.lineplot(
    data=data_melted,
    x='Month', y='Value', hue='Variable', style='Variable',
    markers=True, dashes=False, linewidth=2.5
)

# Customize the plot
plt.title("Average Monthly Temperature and Rainfall", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Value (Â°C / mm)")
plt.legend(title="Variable", loc="upper right")
sns.despine()

plt.show()
```
