Merge

```python

import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['B', 'C', 'D'],
    'value2': [4, 5, 6]
})

# Merge on 'key' column (default is inner join)
merged_df = pd.merge(df1, df2, on='key')

print(merged_df)
# Inner join (only matching keys)
inner_merge = pd.merge(df1, df2, on='key', how='inner')

# Left join (all keys from df1)
left_merge = pd.merge(df1, df2, on='key', how='left')

# Right join (all keys from df2)
right_merge = pd.merge(df1, df2, on='key', how='right')

# Outer join (all keys from both)
outer_merge = pd.merge(df1, df2, on='key', how='outer')

print("Inner Merge:\n", inner_merge)
print("Left Merge:\n", left_merge)
print("Right Merge:\n", right_merge)
print("Outer Merge:\n", outer_merge)
```

Joins

```python
import pandas as pd

df1 = pd.DataFrame({'value1': [10, 20, 30]}, index=['A', 'B', 'C'])
df2 = pd.DataFrame({'value2': [100, 200, 300]}, index=['B', 'C', 'D'])

# Left join (default)
left_join = df1.join(df2, how='left')

# Right join
right_join = df1.join(df2, how='right')

# Outer join
outer_join = df1.join(df2, how='outer')

# Inner join
inner_join = df1.join(df2, how='inner')

print("Left join:\n", left_join)
print("\nRight join:\n", right_join)
print("\nOuter join:\n", outer_join)
print("\nInner join:\n", inner_join)
```

Pivot Table

```python
import pandas as pd

# Sample data
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 250]
}

df = pd.DataFrame(data)

# Create pivot table
pivot = pd.pivot_table(df,
                       values='Sales',    # column to aggregate
                       index='Date',      # rows
                       columns='Category',# columns
                       aggfunc='sum',     # aggregation function
                       fill_value=0)      # fill NaN with 0

print(pivot)
```
