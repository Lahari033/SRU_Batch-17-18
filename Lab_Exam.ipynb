```python
import matplotlib.pyplot as plt
categories={"Subjects":['Maths','English','Science']}
values={"Marks":[99,100,95]}
plt.bar(categories["Subjects"], values["Marks"], color='Red',width=0.07)
plt.title("Bar Graph")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
categories={"Subjects":['Maths','English','Science']}
values={"Marks":[99,100,95]}
plt.pie(values["Marks"], labels=categories["Subjects"], autopct='%1.1f%%')
plt.title("Pie Chart of Marks by Subject")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
