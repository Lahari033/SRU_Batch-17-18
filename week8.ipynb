```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
data = {
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Chennai'],
    'Temperature(°C)': [32, 28, 26, 33],
    'Rainfall(mm)': [120, 300, 250, 220],
    'Humidity(%)': [60, 85, 80, 75],
    'Latitude': [28.6139, 19.0760, 12.9716, 13.0827],
    'Longitude': [77.2090, 72.8777, 77.5946, 80.2707]
}

df = pd.DataFrame(data)
# Compute correlation matrix
corr = df[['Temperature(°C)', 'Rainfall(mm)', 'Humidity(%)']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - Temperature, Rainfall, Humidity')
plt.show()
# Create a base map centered on India
map_india = folium.Map(location=[22.9734, 78.6569], zoom_start=5, tiles='CartoDB positron')

# Prepare data for heatmap
heat_data = [[row['Latitude'], row['Longitude'], row['Temperature(°C)']] for index, row in df.iterrows()]

# Add heatmap layer
HeatMap(heat_data, radius=25, blur=15, max_zoom=5).add_to(map_india)

# Display map
map_india
```

What additional insights do you gain from combining a
correlation heatmap and a geospatial visualization?


Correlation heatmap tells you how strongly variables are related numerically. For example, it can show that rainfall and humidity move together.

Geospatial visualization shows where these values occur across different cities, giving a spatial context.

Combined insight: You not only know that rainfall and humidity are strongly related, but you also see which cities experience high rainfall and humidity.

Example: Mumbai has both high rainfall (300 mm) and high humidity (85%), while Delhi has lower values (120 mm rainfall, 60% humidity).

This helps link statistical relationships to real-world geographic patterns.

If the correlation between rainfall and humidity is 0.9, what
does it mean?

A correlation of 0.9 indicates a very strong positive relationship.

Interpretation: As rainfall increases, humidity also increases almost proportionally.

In practical terms: cities with more rainfall, like Mumbai, also tend to have higher humidity.

Conversely, cities with low rainfall, like Delhi, have lower humidity.
